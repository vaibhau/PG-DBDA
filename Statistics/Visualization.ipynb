# Difference between the Bar Chart and Histogram

### Bar Chart

- x axis variable has categorical data.
- In bar charts, because it has categorical data on x-axis variable. Hence bars will not touch with each other.
- y-axis value will always counted.
- width of the bar graph is same / cannot be change.

### Histogram

- x axis variable has continuous data.
- In histogram, because it has continuous data on x-axis variable. Hence bars will touch with each other.
- y-axis value will always counted.
- width of the histogram can be change.

```python
counts = [34,56,28,40]
languages = ['Marathi','Hindi','English','Kannada']
plt.bar(languages,counts)
```

```python
counts = [34,56,28,40]
languages = ['Marathi','Hindi','English','Kannada']
plt.pie(counts,labels=languages)
```

```python
counts = [34,56,28,40]
languages = ['Marathi','Hindi','English','Kannada']
plt.barh(languages,counts)
```
