# Clustering Algorithm : DBSCAN Algorithm

## Importing packages

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

## Loading dataset

```python
df = pd.read_csv('D14data1.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
#input data
x = df.iloc[:,[3,4]].values
x
```

## Feature Scaling

```python
# from sklearn.preprocessing import StandardScaler
# sc = StandardScaler()
# x = sc.fit_transform(x)
```

## Model Building

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=3, min_samples=4, metric='euclidean')
cls = dbscan.fit(x)
```

```python
la = cls.labels_
la
```

## Prediction

```python
y_pred = dbscan.fit_predict(x)
y_pred
```

```python
plt.scatter(x[y_pred == 0,0],x[y_pred == 0,1],s=50, c='red', label = 'C1')
plt.scatter(x[y_pred == 1,0],x[y_pred == 1,1],s=50, c='blue', label = 'C2')
plt.scatter(x[y_pred == 2,0],x[y_pred == 2,1],s=50, c='green', label = 'C3')
plt.scatter(x[y_pred == 3,0],x[y_pred == 3,1],s=50, c='cyan', label = 'C4')
plt.scatter(x[y_pred == 4,0],x[y_pred == 4,1],s=50, c='orange', label = 'C5')
plt.scatter(x[y_pred == 5,0],x[y_pred == 5,1],s=50, c='yellow', label = 'C6')
plt.scatter(x[y_pred == 6,0],x[y_pred == 6,1],s=50, c='pink', label = 'C7')
plt.scatter(x[y_pred == 7,0],x[y_pred == 7,1],s=50, c='grey', label = 'C8')
plt.scatter(x[y_pred == 8,0],x[y_pred == 8,1],s=50, c='purple', label = 'C9')
plt.title('DBSCAN Clustering')
plt.xlabel('Customer Score')
plt.ylabel('Annual Income')
```
