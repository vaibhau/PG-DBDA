# Decision Tree - Multi Class Classification

## Importing packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

# Loading dataset from sklearn library

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
type(iris)
```

## Feature

```python
x = iris.data
x
```

## Target

```python
y = iris.target
y
```

## Splitting data into training and testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25, random_state = 0)
```

## Model building

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
```

```python
classifier.fit(x_train,y_train)
```

## Prediction

```python
y_pred = classifier.predict(x_test)
y_pred
```

Visualization

```python
from sklearn import tree
text = tree.export_text(classifier)
print(text)
```

```python
tree.plot_tree(classifier)
```

```python
from sklearn import tree
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(classifier,feature_names = iris.feature_names,class_names = "Class label",filled = True)
```

## Evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

### Confusion Matrix

```python
confusion_matrix(y_test,y_pred)
```

### Classification Report

```python
print(classification_report(y_test,y_pred))
```

### Accuracy Report

```python
print(accuracy_score(y_test,y_pred))
```
