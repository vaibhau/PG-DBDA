# Tuple

- Tuples are used to store multiple values in a single variable.
- Tuple is one of the 4 built-in data types in a python used to store collections of data.
- A tuple is a collection which is ordered and unchangeable.
- Tuples are denoted with round brackets.


```python
vehicles = ('car','bike','truck','jeep','bus')
vehicles
```

## Tuple Items

- Tuple items are ordered, changeable and allow duplicate values.
- Tuple items are indexed, the first item has index [0] and so on...

## Ordered

- When we say tuples are ordered, it means that the items have the items have a defined order, and that order will not change.


## Unchangeable

- Tuples are unchangeable means that we cannot change, add or remove items after the tuple has been created.

## Allow Duplicates

- tuples are indexed, so they can have items with same value.


```python
vehicles = ('car','bike','truck','jeep','bus','truck','jeep','scooty','tempo','train','truck','jeep')
print(vehicles)
```

## Tuple Length

- To get how many items a tuple has, we used len() function.

```python
len(vehicles)
```

## Create a Tuple with one item

- To recognize a tiple with only one item, you have to add a comma after the item, otherwise python will not recognize it as a tuple.

```python
# it will recognize as a string
thistupple = ('apple')
thistupple
```

```python
type(thistupple)
```

```python
thistupple = ('apple',)
thistupple
```

```python
type(thistupple)
```

## Tuple Items - Data Type

- Tuples can store values with any data type.

```python
thistuple = ('vaibhav',73,'mumbai',True,'male')
thistuple
```

## Type

- From Python's perspective, tuples are defined as objects with the data type 'tuple'

```python
print(type(thistuple))
```

## The tuple() Constructor

- It is also possible to use tuple constructor to make a tuple.


```python
tuple1 = tuple(('vaibhav','dbda',73))
tuple1
```
